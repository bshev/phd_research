```python
import h5py as h5
import numpy as np
```

```python
#parser to convert Casino simulated electron trajectories and energies to h5 format

files = ["data/hBN_100nm_Si_2kV.dat","data/hBN_100nm_Si_5kV.dat","data/hBN_100nm_Si_10kV.dat"]
for file in files:
    input_file = open(file)
    beam_energy = file.split('_')[-1]
    beam_energy = beam_energy.replace('.dat','')

    write_name = file.replace('.dat','.h5')
    with h5.File(write_name,'w') as h5f:
        h5f.attrs['Columns'] = ['X','Y','Z','Energy']
        h5f.attrs['BeamEnergy'] = beam_energy
        
        n=-1
        found_header = False
        output=[]
        for line in input_file:
            if line == 'X\tY\tZ\tRx\tRy\tRz\tEnergy\tRegion\n':
                electron_data = []
                found_header = True
                n+=1
    
            elif '-------' in line:
                if found_header:
                    output.append(electron_data)
    
                    dataset = h5f.create_dataset(f"electron_{n}",data=electron_data)
                    #dataset.attrs['Columns'] = ['X','Y','Z','Energy']
                
                found_header=False
        
            elif found_header:
                line_item = line.strip().split('\t')
                x,y,z = float(line_item[0]),float(line_item[1]),float(line_item[2])
                Energy = float(line_item[6])
                electron_data.append([x,y,z,Energy])
```
